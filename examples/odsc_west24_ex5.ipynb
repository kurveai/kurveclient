```python
# We will create a number of dimensions and targets
# customer ordering in the next 30 days
# customer ordering in the next 45 days
# customer interacting with a notification in the next week
# notification-level interaction likelihood within 3 days
```

```python
cut_date = datetime.datetime(2023, 5, 6)

params = [
    {'path': p.absolute(),
     'parent_node': f'{p.absolute()}/cust.csv', 
     'label_node':f'{p.absolute()}/orders.csv', 
     'cut_date':cut_date,
    'compute_period':365,
    'label_period':30,
     'storage_format': 'csv'
    },


    {'path':p.absolute(),
    'parent_node': f'{p.absolute()}/cust.csv', 
     'label_node':f'{p.absolute()}/orders.csv', 
     'cut_date':cut_date,
    'compute_period':365,
    'label_period':45,
    'storage_format': 'csv'
    },

    {'path':p.absolute(),
    'parent_node': f'{p.absolute()}/cust.csv', 
     'label_node':f'{p.absolute()}/orders.csv', 
     'cut_date':cut_date,
    'compute_period':365,
    'label_period':90,
    'storage_format': 'csv'
    },

    {'path':p.absolute(),
    'parent_node': f'{p.absolute()}/cust.csv', 
     'label_node':f'{p.absolute()}/orders.csv', 
     'cut_date':cut_date,
    'compute_period':365,
    'label_period':90,
    'storage_format': 'csv'
    },


    {'path':p.absolute(),
    'parent_node': f'{p.absolute()}/cust.csv', 
     'label_node':f'{p.absolute()}/notifications.csv', 
     'cut_date':cut_date,
    'compute_period':365,
    'label_period':3,
    'storage_format': 'csv'
    },

    {'path':p.absolute(),
    'parent_node': f'{p.absolute()}/notifications.csv', 
     'label_node':f'{p.absolute()}/notification_interactions.csv', 
     'cut_date':cut_date,
    'compute_period':365,
    'label_period':3,
    'storage_format':'csv'
    },


]
```

```python
vects = []
for kwargs in params:
    vect = autofe_local_source(**kwargs)
    vects.append(vect)
for vect in vects:
    print(f'finished building feature vector with shape {vect.shape}')
```
